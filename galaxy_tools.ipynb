```python
import bioblend.galaxy

def get_tools(server, api_key):
    gi = bioblend.galaxy.GalaxyInstance(url=server, key=api_key)
    gi.tools.get_tools()
    tools = gi.tools.get_tools()
    sections=set()
    for element in tools:
        sections.add(element['panel_section_name'])
    sorted_sections = list(sections)
    sorted_sections.sort()
    #print sorted_sections
    for section in sorted_sections:
        for element in tools: 
            if element['panel_section_name']==section:
                print section,'\t;', element['name']
                
get_tools('https://usegalaxy.org/', '29197b9c690d49fbf511447d4a676ebc')
```
